# Importing Libraries


```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

#  Reading Data

```python

# loading the csv data to a Pandas DataFrame
gold_data = pd.read_csv('gld_price_data.csv')
```

# Correlation 

```python

correlation = gold_data.corr()
```

```python

# constructing a heatmap to understand the correlation
plt.figure(figsize = (8,8))
sns.heatmap(correlation,  fmt='.1f',annot=True, annot_kws={'size':8}, cmap='Blues')
#first parameter is the data.
#second parameter is a formatting code to make outputs in float number.
#third parameter 'True' to write the data value in each cell.
#fourth parameter is the size of each cell
#fifth parameter indicate to the color of cells.
```

```python

# correlation values of GLD
print(correlation['GLD'])
```

```python

# checking the distribution of the GLD Price
sns.distplot(gold_data['GLD'],color='green',kde=False)
```

```python

X = gold_data.drop(['Date','GLD'],axis=1)
Y = gold_data['GLD']
```

```python

print(X)
```

```python

print(Y)
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() 
X= scaler.fit_transform(X)
print(X)
```

# Splitting Data into Training Set and Test Set

```python

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

```python

# Training the model
regression = LinearRegression()
regression.fit(X_train, Y_train)
```

```python

# prediction on Test Data
test_data_prediction = regression.predict(X_test)
```

```python

df = pd.DataFrame({'Actual':Y_test,'predicted':test_data_prediction}) # comparing between predicted value and actual value.
print(df)
```

```python

# R squared error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared error : ", error_score)
```

```python

Y_test = list(Y_test)
plt.figure(figsize=(10,10))
plt.plot(Y_test, color='red', label = 'Actual Value')
plt.plot(test_data_prediction, color='blue', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()


plt.show()
```
